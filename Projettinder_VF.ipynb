Observe the dataset to understand it. Observe the columns to identify which ones are important and which ones are less important.

```python
# Import and read dataset 
# Used the 'encoding='latin-1' to specify the correct encoding when reading my file because there was likely a character in the data that cannot be decoded using the UTF-8 encoding

dataset = pd.read_csv("Speed_Dating_Data.csv", encoding='latin-1')
dataset.head()
```

To clean our dataset, I strated first by dropping the columns that seem not important to our project.
I used the dropna method to do that

```python
# Drop the columns we are not interested in :

dataset= dataset.drop(columns=['id','idg','condtn','position', 'positin1', 'order', 'partner', 'zipcode','income'])
```

```python
print("The shape of the dataset is now :" ,dataset.shape)
print(" We went from 195 columns to 186 columns")
```

We then look for missing values in our dataset to know if we should clean the dataset or not.

```python
pourcentage_missing = (dataset.isna().mean()*100).round(2)
pourcentage_total_missing=(dataset.isna().mean().mean()*100).round(2) 
print("Percentage of missing values is {}%" .format(pourcentage_total_missing))
print("{}% is a big percentage of missing values so we will have to clean that.".format(pourcentage_total_missing))
```

```python
# Deleting the missing values in our dataset : Deleting the columns where the percentage of missing values is more than 50% :

threshold= int(len(dataset)/2)

dataset.dropna(axis=1, thresh=threshold, inplace=True)
```

```python
# Recalculating the percentage of missing values to see the results after cleaning :

pourcentage_missing = (dataset.isna().mean()*100).round(2)
pourcentage_total_missing=(dataset.isna().mean().mean()*100).round(2)

print("{}% is our new percentage of missing values.".format(pourcentage_total_missing))
```

```python
# Basic stats

print("Number of rows : {}".format(dataset.shape[0]))
print()

print("Number of columns : {}".format(dataset.shape[1]))
print()

print("Percentage of missing values: ")
display(100*dataset.isnull().sum()/dataset.shape[0])

print("Basics statistics: ")
data_desc = dataset.describe(include='all')
display(data_desc)
print()
```

```python
# Display shape of the dataset

print("Number of rows : {}".format(dataset.shape[0]))
print("Number of columns : {}".format(dataset.shape[1]))
```

Let's modify numbers in our dataset into their meaning 

```python
# gender : Female = 0 , Male = 1
dataset['gender'] = dataset['gender'].map({ 0 : "Female", 1 : "Male"})
# match : 1 = yes, 0 = no
dataset['match'] =dataset['match'].map({ 0 : "No", 1 : "Yes"})

# decision : 1 = yes, 0 = no
dataset['dec']  = dataset['dec'].map({ 0 : "No", 1 : "Yes"})

#  Same Race 1= yes, 0=no
dataset['samerace'] = dataset['samerace'].map({ 0 : "No", 1 : "Yes"})

# race : Black/African American = 1 , European/Caucasian-American = 2 , Latino/Hispanic American = 3 , 
# Asian/Pacific Islander/Asian-Americann = 4 , Native American = 5 
dataset["race"] = dataset["race"].apply(lambda x: 'Black/African American' if x == 1
                                  else 'European/Caucasian-American' if x == 2
                                  else 'Latino/Hispanic American' if x == 3
                                  else 'Asian/Pacific Islander/Asian-Americann' if x == 4
                                  else 'Native American' if x == 5
                                   else 'Other'
                                         )

dataset["goal"] = dataset["goal"].apply(lambda x: 'Seemed like a fun night out' if x == 1
                                  else 'To meet new people' if x == 2
                                  else 'To get a date' if x == 3
                                  else 'Looking for a serious relationship' if x == 4
                                  else 'To say I did it' if x == 5
                                   else 'Other'
                                         )
```

### Are shared interests more important than a shared racial background?

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Define the Plotly color palette
color_palette = px.colors.sequential.Blues[::-1]

# Convert Plotly's RGB colors to hex format for Seaborn
hex_palette = [f"#{int(c[0]*255):02x}{int(c[1]*255):02x}{int(c[2]*255):02x}" 
               for c in [tuple(map(lambda x: int(x)/255, color.split('(')[1][:-1].split(','))) 
                         for color in color_palette]]

# Plot using the Seaborn-compatible hex palette
plt.figure(figsize=(12, 6))
plt.title("Race Similarity Importance", fontsize=18)
sns.countplot(x='imprace', data=dataset, palette=hex_palette)
plt.show()
#imprace:How important is it to you (on a scale of 1-10) that a person you date be of the same racial/ethnic background?
```

Most people don't care if they share the same race with their partner.

```python
# Are shared interests more important than a shared racial background?

dataset['samerace_numeric'] = dataset['samerace'].map({'No':0, 'Yes':1 })
dataset['match_numeric'] = dataset['match'].map({'No':0, 'Yes':1 })
correlation= dataset[['shar1_1', 'samerace_numeric', 'match_numeric']].corr()
print(correlation)
```

```python
plt.figure(figsize=(10, 5))
plt.title('Correlation between shared interests, racial and match')
plt.grid(True)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
for i in range(correlation.shape[0]):
    for j in range(correlation.shape[1]):
        plt.text(j + 0.5, i + 0.5, f"{correlation.iloc[i, j]:.2f}", ha='center', va='center', color='black')
sns.heatmap(correlation, cmap='coolwarm' ,vmin=-0.1, vmax=0.2)
plt.show()
```

There is a weak correlation between matches and both shared interests and shared race. But shared race has a stronger correlation with match compared to shared interests.

```python
import plotly.graph_objects as go
import plotly.express as px

# Define color palettes
color_palette0 = px.colors.sequential.Reds[::-1]
color_palette1 = px.colors.sequential.Greens[::-1]

# Map 'No' and 'Yes' to colors
color_discrete_map = {
    'No': color_palette0[0],  # Use the darkest red
    'Yes': color_palette1[0]  # Use the darkest green
}

# Get data for the pie chart
match_df = dataset['match'].value_counts()

# Assign colors based on labels
colors = [color_discrete_map[label] for label in match_df.index]

# Create the Pie chart
trace = go.Pie(
    labels=match_df.index,
    values=match_df.values,
    marker=dict(colors=colors)  # Set colors
)

# Define the layout
layout = go.Layout(
    title={'text': 'Proportion of Yes vs No for a match', 'x': 0.5, 'yanchor': 'top', 'y': 0.97},
    width=700,
    height=500,
    legend=dict(
        x=0.9,
        y=0.3
    )
)

# Create and show the figure
fig = go.Figure(data=[trace], layout=layout)
fig.show()
```

Only 16.5% matches
